```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
sns.countplot(data=df, y="job_type", hue="will_vote", ax=axes[0], palette="cool", 
              order=df['job_type'].value_counts().index)
axes[0].set_title("Job Type vs Will Vote")
axes[0].set_ylabel("Job Type")
axes[0].set_xlabel("Count")
sns.countplot(data=df, y="company_type", hue="will_vote", ax=axes[1], palette="cool", 
              order=df['company_type'].value_counts().index[:10])
axes[1].set_title("Company Type vs Will Vote")
axes[1].set_ylabel("Company Type")
axes[1].set_xlabel("Count")

plt.tight_layout()
plt.show()
```

```python
# insights >> Certain job types (e.g., "commercial organization," "law enforcement agency") show varied voting behavior, with some favoring "definitely yes" responses.
# insights >> Participants from industries like "farming" or "tech, programming, communications" show distinct voting preferences.
```

```python
sns.set(style="whitegrid")

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(df['age'], kde=True, ax=axes[0], color='blue', bins=20)
axes[0].set_title('Distribution of Age')
axes[0].set_xlabel('Age')

sns.histplot(df['weight1'], kde=True, ax=axes[1], color='green', bins=20)
axes[1].set_title('Distribution of Weight1')
axes[1].set_xlabel('Weight1')

plt.tight_layout()
plt.show()

categorical_columns = ['sex', 'federal_district', 'type_of_city', 'candidate']
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

for col, ax in zip(categorical_columns, axes.ravel()):
    sns.countplot(data=df, x=col, ax=ax, palette='viridis', order=df[col].value_counts().index)
    ax.set_title(f'Distribution of {col}')
    ax.set_ylabel('Count')
    ax.set_xlabel(col)
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# insights >> The age distribution is slightly right-skewed, with the majority of participants between 30 and 60 years old.
# insights >> The weighting factor has a distribution centered around 1, with a range from ~0.5 to ~2.5.
# insights >> There is a slight majority of female participants compared to male participants.
# insights >> "Putin" is the most frequent candidate mentioned, followed by others.
# insights >> Most participants are from cities with populations between 100,000 and 500,000.
```
